```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Assignment_4_weight_price.csv")
df
```

```python
x = df[['weight']]
y = df[['price']]
```

```python
mean_x = np.mean(x)
mean_x
```

```python
mean_y = np.mean(y)
mean_y
```

```python
dev_x = x - mean_x
dev_y = y - mean_y
```

## apply:




```python
m = np.sum(dev_x * dev_y) / np.sum(dev_x ** 2)
m
```

```python
# y = mx + c >> c = y - mx >> c = mean_y - m(mean_x)

c = mean_y - m * mean_x
c
```

```python
# Predict price when weight is 6
y = m * 6 + c
y
```

## Compute Residuals for each data point

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(df[['weight']] , df[['price']])
```

```python
# value of m
reg.coef_
```

```python
# value of c
reg.intercept_
```

```python
df['predicted_price'] = reg.predict(x)
df
```

```python
df['residuals'] = df['price'] - df['predicted_price']
df
```

## MSE, MAE calculations

```python
from sklearn.metrics import  mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(df[['predicted_price']], df[['price']])
mse
```

```python
mae = mean_absolute_error(df.predicted_price , df.price)
mae
```

```python
plt.scatter(x,y, color='blue')

plt.plot(x, df.predicted_price)
plt.scatter(x, df.predicted_price, color='red')

plt.xlabel('Weight')
plt.ylabel('Price in Taka')
plt.title('Price of vegetable')
```
