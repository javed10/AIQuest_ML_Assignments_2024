```python
import pandas as pd
import numpy as no
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('UCI_Credit_Card.csv')
df.head()
```

### First check for null values

```python
df.isnull().sum()
```

### NOTE: Rashed changed row 2 & 3. He made null values for Age column to show us how to handle nulls. He used mean()

```python
# for x features drop column default.payment.next.month
x = df.drop(['default.payment.next.month'], axis=1)
```

```python
x.shape
```

```python
y = df['default.payment.next.month']
```

```python
y.shape
```

### Now seperate train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state=42)
```

### Since we will use SVM (support vector machine), we need to import the same from sklearn

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
```

```python
svm_model.fit(xtrain, ytrain)
```

```python
y_predict = svm_model.predict(xtest)
y_predict
```

```python
# check the performance of the model

svm_model.score(xtest, ytest)
```

### Let's try few other classification models to see if the performance improves

```python
# we will use random forest model
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model = RandomForestClassifier()
```

```python
# train the model
rfc_model.fit(xtrain, ytrain)
```

```python
y_rfc_predict = rfc_model.predict(xtest)
```

```python
# performance is better than svm
rfc_model.score(xtest, ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc_model = DecisionTreeClassifier()
```

```python
dtc_model.fit(xtrain, ytrain)
```

```python
# performance not good 
dtc_model.score(xtest, ytest)
```
